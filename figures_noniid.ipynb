```python
import matplotlib.pyplot as plt
import numpy as np

# Data for FL and CL training loss
rounds = list(range(1, 21))

fl_loss = [
    0.09649372818137561, 0.05300187722260964, 0.04285150852873106, 0.035803473881312275,
    0.033079274156921086, 0.031474802849232514, 0.029568596492233527, 0.0286105772598289,
    0.02607978816505021, 0.024749611879632524, 0.025053109273088514, 0.024426539847021568,
    0.021928583124816048, 0.023194190911091966, 0.021561155202493635, 0.02057572119139443,
    0.019819465022830655, 0.019688160961200636, 0.018731341252651378, 0.01907506213757894
]

cl_loss = [
    0.0963, 0.0344, 0.0273, 0.0234, 0.0208, 0.0186, 0.0169, 0.0155, 0.0148, 0.0135,
    0.0131, 0.0126, 0.0122, 0.0113, 0.0111, 0.0108, 0.0106, 0.0098, 0.0099, 0.0097
]

# Calculate RMSE between FL and CL training loss
rmse = np.sqrt(np.mean((np.array(fl_loss) - np.array(cl_loss))**2))

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot FL training loss
ax.plot(rounds, fl_loss, marker='o', linestyle='-', label='FL (iid)', color='blue')

# Plot CL training loss
ax.plot(rounds, cl_loss, marker='s', linestyle='-', label='CL (iid)', color='red')

# Set labels and title with bold font
ax.set_xlabel('Epochs', fontweight='bold')
ax.set_ylabel('Training Loss', fontweight='bold')

# Add RMSE value to the plot
ax.annotate(f'RMSE: {rmse:.4f}', xy=(0.7, 0.1), xycoords='axes fraction', fontsize=12, color='green', fontweight='bold')

# Add a legend
ax.legend()

# Set title with bold font
# ax.set_title('CNN', fontweight='bold')

# Make tick labels on both axes bold
for ticklabel in ax.get_xticklabels():
    ticklabel.set_weight('bold')

for ticklabel in ax.get_yticklabels():
    ticklabel.set_weight('bold')

# Show the plot
plt.grid(True)

# Save the combined plot as a PNG image
plt.savefig('iid_train_loss_cl_fl_cnn.png')

plt.show()

from google.colab import files

# Download the plot
files.download('iid_train_loss_cl_fl_cnn.png')
```
